````python
import random

def roulette_simulation(N):
    wheel = ['red'] * 18 + ['black'] * 18 + ['green'] * 2

    total_earnings = 0

    for _ in range(N):
        result = random.choice(wheel)

        if result == 'black':
            total_earnings += 1
        else:
            total_earnings -= 1

    return total_earnings

N = 1000
earnings = roulette_simulation(N)
print(f"Total earnings after {N} rounds: ${earnings}")
````

### **Overview of the Function**
The function `roulette_simulation(N)` simulates the outcome of betting 1 dollar on black in a game of roulette for \( N \) rounds. It calculates the total earnings after \( N \) rounds, where each round is based on the outcome of the roulette spin.

### **Simulation Process and Logic**
For each round, the outcome is randomly chosen from three possible outcomes: 'red', 'black', and 'green'. The probability of landing on each color is based on the number of slots on the roulette wheel: 18 red slots, 18 black slots, and 2 green slots. If the outcome is 'black', the function adds 1 dollar to the total earnings (indicating a win). Otherwise, the total earnings are reduced by 1 dollar (indicating a loss for red or green). The final total earnings after \( N \) rounds are returned by the function.

````python
def monte_carlo_simulation(N, simulations=100000):
    results = [roulette_simulation(N) for _ in range(simulations)]
    return results

def plot_distribution(N_values):
    plt.figure(figsize=(10, 6))

    for N in N_values:
        results = monte_carlo_simulation(N)
        plt.hist(results, bins=50, alpha=0.5, label=f'N={N}')

    plt.title('Distribution of Total Winnings for Different Values of N')
    plt.xlabel('Total Winnings')
    plt.ylabel('Frequency')

    plt.legend()
    plt.show()

N_values = [10, 25, 100, 1000]
plot_distribution(N_values)
````

### **2. Central Limit Theorem (CLT)**

With $ i = 1, 2, \dots, N $, the **Central Limit Theorem (CLT)** states that if $ N $ is large:

$$
\bar{X} = \frac{1}{N} \sum_{i=1}^{N} X_i = \hat{p} \sim N \left( p, \frac{\hat{p}(1 - \hat{p})}{N} \right)
$$

Where:
- $ \bar{X} $ is the sample mean, representing the proportion of voters supporting the Democratic candidate.
- $ \hat{p} $ is the sample estimate for the proportion of voters supporting the Democratic candidate.
- $ p $ is the true proportion of voters supporting the Democratic candidate.
- $ N $ is the total number of voters (or observations) in the sample.

This means that the sample mean $ \hat{p} $ follows a normal distribution $ N(p, \frac{p(1 - p)}{N}) $, which is centered around the true proportion $ p $ with variance $ \frac{p(1 - p)}{N} $.

---

### **3. Confidence Interval for $ p $**

The Central Limit Theorem (CLT) tells us that, as $ N $ grows larger, the sample mean $ \hat{p} $ becomes approximately normally distributed. Using this result, we can compute a **95% confidence interval** (CI) for $ p $.

- The formula for a 95% confidence interval for $ p $ is:
$$
CI = \hat{p} \pm 1.96 \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{N}}
$$

Where:
- $ \hat{p} $ is the sample estimate for the proportion of voters supporting Clinton.
- $ 1.96 $ is the Z-value corresponding to a 95% confidence level.
- $ N $ is the sample size (the total number of observations).

This formula gives a range within which we are 95% confident the true proportion $ p $ lies. The larger the sample size $ N $, the narrower the confidence interval, indicating a more precise estimate of $ p $.

A Monte Carlo simulation is a computational technique used to understand the impact of uncertainty and variability in mathematical models. It involves running many simulations using random inputs to estimate the outcome of a system or process. This method helps in estimating probabilities, evaluating uncertainty, and modeling complex situations where traditional analytical methods may be difficult or impossible to apply.

### Key Concepts of Monte Carlo Simulation:

1. **Random Sampling**: Monte Carlo simulations rely on generating random numbers to simulate various possible outcomes. These random numbers are often drawn from a probability distribution that represents the uncertain variables in the model.

2. **Repetition of Simulations**: The simulation is repeated many times (often thousands or millions of times). Each repetition uses different random samples to simulate different possible outcomes. The goal is to build a large number of scenarios to understand the range of possible results and their probabilities.

3. **Estimation**: After running a large number of simulations, the results are analyzed statistically. For example, you might calculate the mean, variance, confidence intervals, or other measures that describe the likely outcomes of the model.

### **1. Hypothesis Test for the Spread \( d \)**

To perform the hypothesis test for whether the spread \( d \) is significantly different from zero, we are conducting the following hypothesis test:

### **Hypotheses**:

- **Null Hypothesis** ($H_0$): $d = 0$ (the spread is zero, meaning no difference in support between Clinton and Trump).
- **Alternative Hypothesis** ($H_a$): $d \neq 0$ (the spread is not zero, indicating a difference in support between Clinton and Trump).

---

### **2. Test Statistic**:

We will use a **Z-test** to test this hypothesis. The test statistic $z$ is given by the formula:

$$
z = \frac{d - 0}{SE_d}
$$

Where:
- $d$ is the observed spread, calculated as $d = 2\hat{p} - 1$ (the estimated difference in support between Clinton and Trump).
- $SE_d$ is the standard error for the spread. Since the spread is $2\hat{p} - 1$, the standard error for the spread is twice the standard error of $\hat{p}$, which is calculated as:

$$
SE_d = 2 \times SE_{\hat{p}}
$$

Where $SE_{\hat{p}}$ is the standard error for $\hat{p}$, calculated using the formula:

$$
SE_{\hat{p}} = \sqrt{\frac{\hat{p}(1 - \hat{p})}{N}}
$$

---

### **3. p-value**:

The p-value is the probability of observing a test statistic as extreme as, or more extreme than, the observed test statistic, under the null hypothesis. Since this is a two-tailed test, we calculate the p-value as:

$$
\text{p-value} = 2 \times (1 - \Phi(|z|))
$$

Where $\Phi$ is the cumulative distribution function (CDF) of the standard normal distribution.

---

### **4. Steps**:

1. **Calculate the Spread** $d$:
   $$
   d = 2 \hat{p} - 1
   $$

2. **Calculate the Standard Error for the Spread**:
   $$
   SE_d = 2 \times SE_{\hat{p}}
   $$

3. **Calculate the Test Statistic**:
   $$
   z = \frac{d}{SE_d}
   $$

4. **Calculate the p-value**:
   $$
   \text{p-value} = 2 \times (1 - \Phi(|z|))
   $$

5. **Decision Rule**:
   - If the p-value is less than the significance level (typically 0.05), reject the null hypothesis ($H_0$).
   - If the p-value is greater than 0.05, fail to reject the null hypothesis.

### **Performing an Independent t-test**:

The independent t-test is used to determine if there is a statistically significant difference between the means of two independent groups (in this case, the Drug group and the Placebo group). The formula for the t-statistic is:

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

Where:
- $ \bar{x}_1 $ and $ \bar{x}_2 $ are the sample means for each group (Drug and Placebo).
- $ s_1^2 $ and $ s_2^2 $ are the sample variances.
- $ n_1 $ and $ n_2 $ are the sample sizes.

---

### **Interpreting the Results**:

- After performing the t-test, we interpret the **p-value**:
  - If the **p-value** is less than the significance level (typically 0.05), we **reject the null hypothesis** and conclude that there is a significant difference between the groups.
  - If the **p-value** is greater than 0.05, we **fail to reject the null hypothesis** and conclude that there is no significant difference between the groups.

---

### **Significance Level**:

- The **p-value** is typically set to **0.05** (5%), which means that if the p-value is smaller than 0.05, we consider the difference to be statistically significant.
- If the **p-value** is set to a **higher level** (e.g., 0.1), the threshold for rejecting the null hypothesis becomes less stringent, and some tests that were not significant at 0.05 might become significant.

---

### **Scipy Library** for t-tests:

- The **`alternative`** argument specifies the type of alternative hypothesis:

- The **`equal_var`** argument specifies whether to assume equal variance between the two groups:
